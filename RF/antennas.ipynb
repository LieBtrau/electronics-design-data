# Poynting vector
The Poynting vector represents the power per unit area carried by an electromagnetic wave.
$$|S| = \frac{P_{t}*G_{t}}{4*\Pi*d^{2}} $$

Conditions:
- The antenna is in free space (no obstacles or reflections)

Where:
- $|S|$ is the magnitude of the Poynting vector (power per unit area)
- $P_{t}$ is the transmission power
- $G_{t}$ is the gain of the transmission antenna
- $d$ is the distance between the transmitting and receiving antennas
- $Z_{0}$ is the impedance of free space (approximately 377 ohms)

An alternative expression for the Poynting vector is, given by the cross product of the electric field (E) and the magnetic field (H):
$$|S| = |\vec{E} \times \vec{H}| = \frac{|\vec{E}|^2}{Z_{0}}$$

Conditions:
- Far field region (where the distance from the antenna is much greater than the wavelength of the signal)

Where:
- $|\vec{E}|$ is the magnitude of the electric field
- $|\vec{H}|$ is the magnitude of the magnetic field

# Generalized gain & area relationship
$$
G=\frac{4*\Pi*A_{eff}}{\lambda^{2}}
$$
Where:
- $G$ = gain of the antenna
- $A_{eff}$ = effective area of the antenna
- $\lambda$ = wavelength of the signal

The effective area is derived from the physical area ($A_{phys}$) multiplied by the aperture efficiency ($\epsilon_a$), expressed as $A_{eff} = \epsilon_a A_{phys}$. Typical aperture efficiencies range from 0.35 to over 0.70.

```python
## Calculate effective area given the frequency and the gain of the antenna
import math
f = 1e9 # Hz
G = 1.64  # half-wave dipole antenna gain (dimensionless)

c = 3e8 # m/s
lambda_ = c / f # m

A_eff = (lambda_**2) * G/ (4 * math.pi)
print(f"The effective area of the receiving antenna is approximately {A_eff:.2e} m^2")
```

## E-field strength at the receiver
The electric field strength at the receiver can be calculated using the Poynting vector and the impedance of free space:
$$|\vec{E}| = \sqrt{S * Z_{0}} = \sqrt{\frac{P_{t}*G_{t}}{4*\Pi*d^{2}} * Z_{0}}$$

Where:
- $|\vec{E}|$ = electric field strength at the receiver
- $S$ = magnitude of the Poynting vector
- $Z_{0}$ = impedance of free space (approximately 377 ohms)
- $P_{t}$ = transmission power
- $G_{t}$ = gain of the transmission antenna
- $d$ = distance between the transmitting and receiving antennas

```python
# What is the E-field strength at the receiver given the transmission power, the gain of the transmitting antenna, and the distance between the transmitting and receiving antennas?
import math
Pt = 1.17e-8 # W
Gt = 1.6 # dimensionless, for a half-wave dipole antenna
d = 3 # m
Z0 = 120 * math.pi # ohms

E = math.sqrt((Pt * Gt) / (4 * math.pi * d**2) * Z0)
print(f"The electric field strength at the receiver is approximately {E:.2e} V/m")
```

# Calculating the received power
$$
P_{r}=|S|*A_{eff}
$$

For parabolic antennas and 2D-phased arrays, the A<sub>eff</sub> is a physical property of the antenna, so this form of the formula is useful.  

For dipole antennas, the A<sub>eff</sub> is frequency dependent, while the gain is a physical property of the antenna, so the contemporary version of the Friis equation is more suitable.

$$
P_{r}=P_{t}*G_{t}*G_{r}*\left(\frac{\lambda}{4*\Pi*d}\right)^2
$$

Where:
- $P_{r}$ = received power
- $P_{t}$ = transmission power
- $G_{t}$ = gain of the transmission antenna
- $G_{r}$ = gain of the receiving antenna
- $d$ = distance between TX- and RX-antenna
- $\lambda$ = wavelength of the signal

```python
# Given the electric field strength at the receiver and the effective antenna area of the receiving antenna, what is the power induced in the receiving antenna?
A_eff = 0.13 # m^2, effective area of the receiving antenna
P_induced = (E**2 / Z0) * A_eff
print(f"The power induced in the receiving antenna is approximately {P_induced:.2e} W")
# Convert the power to dBm
P_induced_dBm = 10 * math.log10(P_induced / 1e-3)
print(f"The power induced in the receiving antenna is approximately {P_induced_dBm:.2f} dBm")
```

## Required transmission power for a given E-field strength at a given distance from the transmitter
$$P_{t} = \frac{4*\Pi*d^{2}*|\vec{E}|^2}{G_{t}*Z_{0}}$$

Where:
- $P_{t}$ = required transmission power
- $d$ = distance from the transmitter
- $|\vec{E}|$ = desired electric field strength at the receiver
- $G_{t}$ = gain of the transmission antenna
- $Z_{0}$ = impedance of free space (approximately 377 ohms)

```python
# What is the TX-power given a certain E-field strength and distance to the transmitting antenna? 

import math
d = 3 # distance in m
E = 250e-6 # E-field strength in V/m
Gt = 1.6 # gain of the transmitting antenna (for a dipole antenna)
# ---

Z0 = 120 * math.pi # impedance of free space in ohms 
Pt = (4 * math.pi * d**2 * E**2) / (Gt * Z0)
print(f"Required transmission power: {Pt:.2e} W")
# Converting to dBm
Pt_dBm = 10 * math.log10(Pt * 1e3) # converting W to mW for dBm calculation
print(f"Required transmission power: {Pt_dBm:.2f} dBm")
```
